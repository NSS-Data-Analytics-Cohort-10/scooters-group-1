```python
print(trips_table.dtypes)
```

```python
trips_table['day'] = trips_table['startdate'].dt.day
```

```python
trips_per_day_per_scooter = trips_table.groupby(['startdate', 'sumdid']).size().reset_index(name='num_trips')
```

```python
trips_per_day_per_scooter
```

```python
trips_per_day_per_scooter.sort_values(by=['num_trips'], ascending = False)
```
